```python
import functions.properties.client as mongo
import functions.search_place as sp

def apply_search_place(latitude, longitude, keyword, category):
    data_json = sp.search_place(latitude, longitude, keyword, category)
    i = 3 if len(data_json) > 3 else len(data_json)
    places = [{'name': place['name'],
               'coordinates': place['geometry']['location'],
               'types': place['types']
              } for place in data_json[:i]]
    mongo.db.searched_places.insert_many(places)
```

```python
apply_search_place(latitude, longitude, keyword, category)
```
