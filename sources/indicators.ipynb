# Global Entrepreneurship Index Ranking (GEI)
#### The GEI is an economic activity index compiled by US-based The Global Entrepreneurship and Development Institute, which looks at how individual countries across the world allocate resources to promoting entrepreneurship, if indeed they do. The Institute was founded by economic scholars from the Imperial College London, the LSE and the University of PÃ©cs (Hungary) to produce the GEI, which gives a rating of an individual country's entrepreneurship ecosystem. (Wikipedia, 2020)

```python
df_gei = pd.read_csv('../data/global-entrepreneurship-index.csv',
                    index_col=[0])
```

```python
df_gei.head()
```

```python
plt.style.use('seaborn')
fig = plt.figure(figsize=(12,8)) # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

width = 0.35


df_gei.plot.bar(x='country', y='GEI', color='red',
                ax=ax, width=width, position=1, rot=65)
df_gei.plot.bar(x='country', y='GDP', color='blue',
                ax=ax2, width=width, position=0)
ax2.grid(None)
ax.legend(bbox_to_anchor=(0.993,1))
ax2.legend(bbox_to_anchor=(1,0.95))
plt.xlim(-.6)
ax.set_title('Global Entrepreneurship Index (GEI)')
ax.set_xlabel('Country')
ax.set_ylabel('GEI')
ax2.set_ylabel('GDP')
plt.xticks(rotation=55)
fig.tight_layout()
fig.subplots_adjust(top=0.95)
plt.savefig('../outputs/gei-and-gdp-per-country.png')

plt.show()
```
